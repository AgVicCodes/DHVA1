### Step 1:
Load, train, and predict on the original dataset (without applying pre-processing techniques)

```python
# Load required packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

```python
# Load the data
mtcars = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')
mtcars = mtcars.drop('model', axis=1) # Drop the 'model' column
mtcars.info()
```

```python
# Make predictions for 'mpg' from original dataset:

# Split the data into train and test
train = mtcars.sample(frac=0.7, random_state=123)
test = mtcars.drop(train.index)

# Define the model
model = RandomForestRegressor(random_state=123)

# Train the model on original data
model.fit(train.drop('mpg', axis=1), train['mpg'])

# Make predictions
predictions = model.predict(test.drop('mpg', axis=1))

# Calculate MSE
mse = mean_squared_error(test['mpg'], predictions)
print(f'MSE without pre-processing: {mse}')
```

```python
# Plot prediction from original dataset
df_original = pd.DataFrame({'actual': test['mpg'], 'predicted': predictions, 'Version': 'Without pre-processing'}) # Create a dataframe to plot
plt.figure(figsize=(7, 7))
sns.scatterplot(data=df_original, x='actual', y='predicted')
plt.plot([min(df_original['actual']), max(df_original['actual'])], [min(df_original['actual']), max(df_original['actual'])]) # plot one-to-one line for reference
plt.title('Predictions without pre-processing')
plt.show()
```
